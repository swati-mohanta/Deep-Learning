```python
import cv2
import matplotlib.pyplot as plt
img1 = cv2.imread('Image_1.jpg')
img2 = cv2.imread('Image_2.jpg')
img1 = cv2.resize(img1, (400, 400))
img2 = cv2.resize(img2, (400, 400))
add = cv2.add(img1, img2) # Addition
subtract = cv2.subtract(img1, img2) # Subtraction
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
add_rgb = cv2.cvtColor(add, cv2.COLOR_BGR2RGB)
sub_rgb = cv2.cvtColor(subtract, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.imshow(img1_rgb)
plt.title("Image 1")
plt.axis('off')
plt.subplot(2, 2, 2)
plt.imshow(img2_rgb)
plt.title("Image 2")
plt.axis('off')
plt.subplot(2, 2, 3)
plt.imshow(add_rgb)
plt.title("Addition Result")
plt.axis('off')
plt.subplot(2, 2, 4)
plt.imshow(sub_rgb)
plt.title("Subtraction Result")
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
img1 = cv2.imread('Image_1.jpg', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('Image_2.jpg', cv2.IMREAD_GRAYSCALE)
if img1.shape != img2.shape:
    height, width = min(img1.shape[0], img2.shape[0]), min(img1.shape[1], img2.shape[1])
    img1 = cv2.resize(img1, (width, height))
    img2 = cv2.resize(img2, (width, height))
_, binary1 = cv2.threshold(img1, 120, 255, cv2.THRESH_BINARY)
_, binary2 = cv2.threshold(img2, 120, 255, cv2.THRESH_BINARY)

bitwise_and = cv2.bitwise_and(binary1, binary2)
bitwise_or = cv2.bitwise_or(binary1, binary2)

plt.figure(figsize=(12, 6))
plt.subplot(2, 3, 1)
plt.imshow(binary1, cmap='gray')
plt.title("Binary Image 1")
plt.axis('off')
plt.subplot(2, 3, 2)
plt.imshow(binary2, cmap='gray')
plt.title("Binary Image 2")
plt.axis('off')
plt.subplot(2, 3, 4)
plt.imshow(bitwise_and, cmap='gray')
plt.title("Bitwise AND")
plt.axis('off')
plt.subplot(2, 3, 5)
plt.imshow(bitwise_or, cmap='gray')
plt.title("Bitwise OR")
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
image = cv2.imread('image1.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, thresh_binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
_, thresh_binary_inv = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)
_, thresh_trunc = cv2.threshold(gray, 127, 255, cv2.THRESH_TRUNC)
_, thresh_tozero = cv2.threshold(gray, 127, 255, cv2.THRESH_TOZERO)
_, thresh_tozero_inv = cv2.threshold(gray, 127, 255, cv2.THRESH_TOZERO_INV)
titles = ["Original Grayscale", "Binary", "Binary Inverted", "Truncate", "To Zero", "To Zero Inverted"]
images = [gray, thresh_binary, thresh_binary_inv, thresh_trunc, thresh_tozero, thresh_tozero_inv]
plt.figure(figsize=(15, 8))
for i in range(6):
    plt.subplot(2, 3, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
image = cv2.imread('image1.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, simple_thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
adaptive_mean = cv2.adaptiveThreshold(
    gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
adaptive_gaussian = cv2.adaptiveThreshold(
    gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
titles = ["Original Grayscale", "Simple Threshold",
          "Adaptive Mean Threshold", "Adaptive Gaussian Threshold"]
images = [gray, simple_thresh, adaptive_mean, adaptive_gaussian]
plt.figure(figsize=(12, 8))
for i in range(4):
    plt.subplot(2, 2, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
image = cv2.imread('image1.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, th1 = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY) # 1. Global thresholding (for comparison)
_, th2 = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU) # 2. Otsu's thresholding
blur = cv2.GaussianBlur(gray, (5, 5), 0) # 3. Otsu's after Gaussian blur
_, th3 = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
images = [gray, th1, th2, th3]
titles = ['Original Grayscale',
          'Global Threshold (127)',
          "Otsu's Threshold",
          "Otsu's + Gaussian Blur"]
plt.figure(figsize=(12, 8))
for i in range(4):
    plt.subplot(2, 2, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
image = cv2.imread('image1.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, mask = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU) 
segmented = cv2.bitwise_and(image, image, mask=mask) 
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
segmented_rgb = cv2.cvtColor(segmented, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15, 6))
plt.subplot(1, 3, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(mask, cmap='gray')
plt.title("Otsu's Threshold (Mask)")
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(segmented_rgb)
plt.title("Segmented Image (Using Mask)")
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
image = cv2.imread('image1.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
avg_blur = cv2.blur(image, (5, 5)) # 1. Average Blur 
gaussian_blur = cv2.GaussianBlur(image, (5, 5), 0) # 2. Gaussian Blur 
median_blur = cv2.medianBlur(image, 5) # 3. Median Blur
avg_rgb = cv2.cvtColor(avg_blur, cv2.COLOR_BGR2RGB)
gaussian_rgb = cv2.cvtColor(gaussian_blur, cv2.COLOR_BGR2RGB)
median_rgb = cv2.cvtColor(median_blur, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('off')
plt.subplot(2, 2, 2)
plt.imshow(avg_rgb)
plt.title("Average Blur")
plt.axis('off')
plt.subplot(2, 2, 3)
plt.imshow(gaussian_rgb)
plt.title("Gaussian Blur")
plt.axis('off')
plt.subplot(2, 2, 4)
plt.imshow(median_rgb)
plt.title("Median Blur")
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
image = cv2.imread('image1.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
constant = cv2.copyMakeBorder(image, 50, 50, 50, 50, cv2.BORDER_CONSTANT, value=(255, 0, 0))
reflect = cv2.copyMakeBorder(image, 50, 50, 50, 50, cv2.BORDER_REFLECT)
replicate = cv2.copyMakeBorder(image, 50, 50, 50, 50, cv2.BORDER_REPLICATE)
constant_rgb = cv2.cvtColor(constant, cv2.COLOR_BGR2RGB)
reflect_rgb = cv2.cvtColor(reflect, cv2.COLOR_BGR2RGB)
replicate_rgb = cv2.cvtColor(replicate, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('off')
plt.subplot(2, 2, 2)
plt.imshow(constant_rgb)
plt.title("Border: Constant (Green)")
plt.axis('off')
plt.subplot(2, 2, 3)
plt.imshow(reflect_rgb)
plt.title("Border: Reflect")
plt.axis('off')
plt.subplot(2, 2, 4)
plt.imshow(replicate_rgb)
plt.title("Border: Replicate")
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
img = cv2.imread('image1.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

sharpen_kernel = np.array([[0, -1, 0],
                           [-1, 5,-1],
                           [0, -1, 0]])
sharpened = cv2.filter2D(img, -1, sharpen_kernel)

plt.figure(figsize=(8,4))
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(sharpened, cv2.COLOR_BGR2RGB))
plt.title("Sharpened")
plt.axis('off')

plt.show()
```
